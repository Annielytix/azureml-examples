```python
storage_options = {"account_name": "azureopendatastorage"}

fs = AzureBlobFileSystem(**storage_options)
fs
```

```python
fs.ls("isdweatherdatacontainer")
```

```python
files = fs.glob(
    "isdweatherdatacontainer/ISDWeather/year=2021/month=1/**.parquet", recursive=True
)
files += fs.glob(
    "isdweatherdatacontainer/ISDWeather/year=2021/month=2/**.parquet", recursive=True
)
files += fs.glob(
    "isdweatherdatacontainer/ISDWeather/year=2021/month=3/**.parquet", recursive=True
)
files += fs.glob(
    "isdweatherdatacontainer/ISDWeather/year=2021/month=4/**.parquet", recursive=True
)
files += fs.glob(
    "isdweatherdatacontainer/ISDWeather/year=2021/month=5/**.parquet", recursive=True
)
files += fs.glob(
    "isdweatherdatacontainer/ISDWeather/year=2020/month=*/**.parquet", recursive=True
)
files = [f"az://{file}" for file in files]
files[-5:]
```

```python
len(files)
```

```python
df = pd.read_parquet(files[0], storage_options=storage_options)
df.head()
```

```python
df.columns
```

```python
columns = ["datetime", "latitude", "longitude", "elevation", "windAngle", "windSpeed", "temperature", "seaLvlPressure", "precipTime", "precipDepth", "stationName", "countryOrRegion",]
```

```python
df = dd.read_parquet(files, storage_options=storage_options, columns=columns, engine="pyarrow")
df = df.set_index(dd.to_datetime(df.datetime).dt.floor("d"), sorted=False).persist()
wait(df)
```

```python
df.describe().compute()
```

```python
places = df.groupby(df.index)[["longitude", "latitude", "elevation"]].mean().compute()
```

```python
plt.figure(figsize=(16, 16))
plt.scatter(places.longitude, places.latitude, c=places.elevation)
plt.title("Lat/long")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid()
plt.colorbar()
```

```python
means = df.groupby(df.index).mean().compute()
means.head()
```

```python
for col in list(means.columns):
    fig = plt.figure(figsize=(16, 8))
    means[col].plot(color="b")
    plt.title("Average of {}".format(col))
    plt.grid()
```
